# ***Read Machine Learning Model***


```
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

#read image
img = cv2.imread("startupschool.png")

#change image too gray scale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

# **Find Faces**

We are going to use the detectMultiscale module from OpenCV. What this does is create a rectangle with coordinates (x,y,w,h) around the face detected in the image. This contains code parameters that are the most important to consider.

```
faces = face_cascade.detectMultiScale(gray,1.1,5)
```

# **Draw rectangle**

```
for (x,y,w,h) in faces:

  # draw rectangle => cv2.rectangle(image,start_point,end_point,color, thickness)
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

imgplot = cv2_imshow(img)
print('Numero de caras en la imagen:',len(faces))
```
